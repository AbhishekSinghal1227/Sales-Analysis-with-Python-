# Exploratory Data Analysis (EDA)

```python
df.columns
```

## EDA on Gender 

```python
sns.countplot(x='Gender', data=df)
```

### Gender Count

```python
ax=sns.countplot(x='Gender', data=df, color='orange')

for bars in ax.containers:
    ax.bar_label(bars,fontsize=10, color='red', padding=1)
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum()
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by ='Amount',ascending=True)
```

### Purchase Power based on Gender

```python
Gender_sales_data =df.groupby(['Gender'],as_index=False)['Amount'].sum()
sns.barplot(x='Gender', y='Amount', data=Gender_sales_data, color= 'orange')
```

### Average Purchase Amount 

```python
Gender_sales_data = df.groupby(['Gender'], as_index=False)['Amount'].mean()
ax= sns.barplot(x='Gender', y='Amount', data=Gender_sales_data, color= 'red')

for bar in ax.containers:
    ax.bar_label(bar, fontsize=10, color='black', padding=3)
```

### From the above graphs we can see that most of the buyers are females and even there purchase power of females is more than men

## EDA on Age

```python
df.columns
```

```python
sns.countplot(x='Age Group', data=df, palette='Set2')  # using palette='Set2' for different colors
```

```python
ax= sns.countplot(x='Age Group',hue='Gender', data=df) 

for bar in ax.containers:
    ax.bar_label(bar)
```

```python
sales_data =df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending =False)
ax=sns.barplot(x='Age Group', y='Amount',data=sales_data,color='yellow')
```

### From all the above graphs we can see the purchase power is maximum in case of Age Gap betweeen 26-35 females

## EDA on State

```python
df.columns
```

## State wise No. of Orders for top 10 states

```python
plt.figure(figsize=(25,7)) 
State_order=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders', ascending =False).head(10)

# Basically we are counting no of orders from each state
ax=sns.barplot(x='State',y='Orders',data=State_order)

for bar in ax.containers:
    ax.bar_label(bar,fontsize=15)
```

## State wise Amount of expenditure for top 10 states

```python
plt.figure(figsize=(25,7)) 
State_amount=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending =False).head(10)

ax=sns.barplot(x='State',y='Amount',data=State_amount,palette='Set2')

```

### From the above graphs we can see that most of the orders and the amount are from UP, Maharashtra and Karnataka respectively

## EDA on Marital Status

```python
ax = sns.countplot(data = df, x = 'Marital_Status',palette='Set2')

for bar in ax.containers:
    ax.bar_label(bar)
```

```python
sales_Maritalstatus = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(data = sales_Maritalstatus, x = 'Marital_Status',y= 'Amount', hue='Gender')
```

### From above graphs we can see that most of the buyers are married (women) and they have high purchasing power as compared to married men

## EDA on Occupation

```python
plt.figure(figsize=(25,7)) 
ax= sns.countplot(data=df, x='Occupation',palette='Set2')
```

```python
plt.figure(figsize=(25,7)) 
sales_Occupation = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(data = sales_Occupation, x = 'Occupation',y= 'Amount',palette='Set2')
```

### From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sectors.

## EDA on Product Category 

```python
plt.figure(figsize=(25,7)) 
ax=sns.countplot(data = df, x='Product_Category',palette='Set2')

for bar in ax.containers:
    ax.bar_label(bar,fontsize=15)
```

```python
plt.figure(figsize=(25,7)) 
sales_ProductCategory = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.barplot(data = sales_ProductCategory, x = 'Product_Category',y= 'Amount',palette='Set2')
```

### From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category
